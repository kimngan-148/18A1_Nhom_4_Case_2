Làm sạch close_price và volume

```python
import pandas as pd
df = pd.read_csv("stock_price.csv")
```

```python
df['close_price'] = pd.to_numeric(
    df['close_price'].astype(str).str.replace(r'[^\d.]', '', regex=True),
    errors='coerce'
)

if 'volume' in df.columns:
    df['volume'] = pd.to_numeric(
        df['volume'].astype(str).str.replace(r'[^\d]', '', regex=True),
        errors='coerce'
    ).fillna(0).astype(int)
```

Chuẩn hóa thông tin sàn và ngành

```python
import pandas as pd

# Giả sử df đã có cột 'exchange' và 'sector'

# 1. Loại bỏ khoảng trắng
df['exchange'] = df['exchange'].str.strip()
df['sector'] = df['sector'].str.strip()

# 2. Chuẩn hóa chữ hoa/chữ thường
df['exchange'] = df['exchange'].str.upper()   # HOSE, HNX, UPCOM
df['sector'] = df['sector'].str.title()       # Viết hoa chữ đầu

# 3. Thống nhất các giá trị phổ biến (nếu có lỗi chính tả)
df['sector'] = df['sector'].replace({
    'Ngan Hang': 'Ngân Hàng',
    'Dien Tu': 'Điện Tử'
})

# 4. Kiểm tra kết quả
print(df[['exchange', 'sector']])
```

Xuất các file sạch để dùng cho các nhiệm vụ sau

```python
import pandas as pd
import os

output_folder = "../data_sạch/"
os.makedirs(output_folder, exist_ok=True)
```

```python
df_info = pd.read_csv("stock_info.csv")
df_price = pd.read_csv("stock_price.csv")
df_trans = pd.read_csv("transactions.csv")
```

```python
if 'exchange' in df_info.columns:
    df_info['exchange'] = df_info['exchange'].str.strip().str.upper()
if 'sector' in df_info.columns:
    df_info['sector'] = df_info['sector'].str.strip().str.title()

if 'close_price' in df_price.columns:
    df_price['close_price'] = pd.to_numeric(
        df_price['close_price'].astype(str).str.replace(r'[^\d.]', '', regex=True),
        errors='coerce'
    )
```

```python

if 'volume' in df_trans.columns:
    df_trans['volume'] = pd.to_numeric(
        df_trans['volume'].astype(str).str.replace(r'[^\d]', '', regex=True),
        errors='coerce'
    ).fillna(0).astype(int)
```

```python


df_info.to_csv(os.path.join(output_folder, "stock_info_clean.csv"), index=False)
df_price.to_csv(os.path.join(output_folder, "stock_price_clean.csv"), index=False)
df_trans.to_csv(os.path.join(output_folder, "transactions_clean.csv"), index=False)

print("Đã xuất các file sạch vào folder data_sạch!")
```
