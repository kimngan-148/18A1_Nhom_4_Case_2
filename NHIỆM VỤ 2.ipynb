```python
# ================================
# NHIỆM VỤ 2 – TRUY VẤN & THỐNG KÊ
# ================================

import pandas as pd

# ----------------
# 1. TẠO DỮ LIỆU GỐC stock_info
# ----------------
data = {
    "ticker": ["vic", "vhm", "VNM", "HPG", "FPT ", "mwg", "vcb", "MBB", "bid", "SSI "],
    "exchange": ["upcom", "UPCoM", "UPCoM", "hose", "upcom", "HOSE", "HNX", "UPCoM", "UPCoM", "UPCoM"],
    "sector": [
        "  tai chinh ",
        "Công nghệ",
        "Bán lẻ",
        "  tai chinh ",
        "Bán lẻ",
        "Bán lẻ",
        "Bán lẻ",
        "Tài chính",
        "  tai chinh ",
        "  tai chinh "
    ]
}

stock_info = pd.DataFrame(data)

# ----------------
# 2. LÀM SẠCH DỮ LIỆU
# ----------------
stock_info["ticker"] = stock_info["ticker"].str.upper().str.strip()
stock_info["exchange"] = stock_info["exchange"].str.upper().str.strip()
stock_info["sector"] = stock_info["sector"].str.strip().str.title()

print("=== STOCK INFO SAU KHI LÀM SẠCH ===")
print(stock_info)

# ----------------
# 3. TẠO DỮ LIỆU stock_price (có giá thiếu)
# ----------------
stock_price = pd.DataFrame({
    "ticker": ["VIC", "VHM", "VNM", "HPG", "MWG"],
    "close_price": [85000, None, 90000, 28000, None]
})

print("\n=== STOCK PRICE ===")
print(stock_price)

# ----------------
# 4. TẠO DỮ LIỆU transactions
# ----------------
transactions = pd.DataFrame({
    "ticker": ["VIC", "VHM", "VNM", "HPG", "MWG"],
    "volume": [10000, 15000, 12000, 18000, 16000]
})

print("\n=== TRANSACTIONS ===")
print(transactions)

# =====================================================
# NHIỆM VỤ 2 – TRUY VẤN & THỐNG KÊ MÔ TẢ
# =====================================================

# 1. LỌC DỮ LIỆU THEO SÀN GIAO DỊCH (VD: UPCOM)
upcom_stocks = stock_info[stock_info["exchange"] == "UPCOM"]

print("\n--- CỔ PHIẾU THUỘC SÀN UPCOM ---")
print(upcom_stocks)

# 2. THỐNG KÊ SỐ LƯỢNG MÃ THEO TỪNG NGÀNH
count_by_sector = stock_info.groupby("sector")["ticker"].count()

print("\n--- SỐ LƯỢNG MÃ THEO NGÀNH ---")
print(count_by_sector)

# 3. TÌM CÁC MÃ CÓ GIÁ ĐÓNG CỬA BỊ THIẾU
missing_price = stock_price[stock_price["close_price"].isna()]

print("\n--- MÃ CÓ GIÁ ĐÓNG CỬA BỊ THIẾU ---")
print(missing_price)

# 4. THỐNG KÊ KHỐI LƯỢNG GIAO DỊCH TRUNG BÌNH
avg_volume = transactions["volume"].mean()

print("\n--- KHỐI LƯỢNG GIAO DỊCH TRUNG BÌNH ---")
print(avg_volume)

# 5. TÌM MÃ CÓ TRONG stock_info NHƯNG KHÔNG CÓ TRONG stock_price
missing_in_price = stock_info[
    ~stock_info["ticker"].isin(stock_price["ticker"])
]

print("\n--- MÃ CÓ TRONG stock_info NHƯNG KHÔNG CÓ TRONG stock_price ---")
print(missing_in_price)
```
